```python
import numpy as np
```

```python
def maxpool2d(input, pool_size):
    N, H, W, C = input.shape
    p_h, p_w = pool_size

    out = np.zeros((N, H-p_h+1, W-p_w+1, C))
    for i in range(out.shape[1]):
        for j in range(out.shape[2]):
            out[:, i, j, :] = np.max(input[:, i:i + p_h, j:j + p_w, :], axis=(1, 2))
    return out
```

```python
input = np.random.randn(10, 128, 128, 64)
maxpool2d(input, (2, 2))
```
